```python
#Fill the NaN value
jobs[2] = jobs[2].fillna(value='unknown')
jobs[2]
```

```python
#Capitalization
jobs['job_name'] = jobs[0].str.title()
jobs['job_name']
```

```python
#Seperate the location and area information 
jobs['location'] = jobs[3].str.split('area:').str[0]
jobs['location'].isna().sum()
```

```python
#Seperate area, with null value
jobs['area'] = jobs[3].str.split('area:').str[1]
jobs['area'].isna().sum()
```

```python
#Remove "Location: "
jobs['location'] = jobs['location'].str.replace("location: ", "")
jobs
```

```python
#Remove ',' in location
jobs[jobs['location'].str.contains(',')]
jobs['location'] = jobs['location'].str.split(',', n=1, expand = True)[0]
jobs['location']
```

```python
#location Dedupilication 
def deduplication(x):
    trim = x.strip()#检查有无空格3
    index = (trim+trim).find(trim,1)#从第二位开始查重trim的index
    if trim != -1:#如果能找到
        return trim[index:]
    else:
        return trim
```

```python
jobs['location']=jobs['location'].apply(deduplication)
```

```python
#remove ‘，’
jobs['area'] = jobs['area'].str.split(',', n=1, expand = True)[0]
jobs['area'] = jobs['area'].fillna('unknown')
jobs['area']
# jobs['area'].str.contains(',').value_counts()
```

```python
jobs['area'].isna().sum()
```

```python
#Use same dedupilication method
jobs['area']=jobs['area'].apply(deduplication)
```

```python
#Post time 
jobs[4] = jobs[4].str.split('ago', n=1,expand=True)[0]
jobs[4] 
```

```python
#get post date
def get_postdate(i):
    today_date = datetime.today()
    x = i.strip()
    if x[len(x)-1]in ['hour', 'hours', 'hr', 'hrs', 'h']:
        date = today_date - relativedelta(hours=int(x[:len(x)-1]))
        return str(date.date().isoformat())
    elif x[len(x)-1]in ['day', 'days', 'd']:
        date = today_date - relativedelta(days=int(x[:len(x)-1]))
        return str(date.date().isoformat())
    else:
        return 'unknown'
```

```python
jobs['post_date'] = jobs[4].apply(get_postdate)
# jobs[4][2]
```
