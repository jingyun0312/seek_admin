```python
#location Dedupilication 
def deduplication(x):
    trim = x.strip()#检查有无空格3
    index = (trim+trim).find(trim,1)#从第二位开始查重trim的index
    if trim != -1:#如果能找到
        return trim[index:]
    else:
        return trim
```

```python
#get post date
def get_postdate(i):
    today_date = datetime.today()
    x = i.strip()
    if x[len(x)-1]in ['hour', 'hours', 'hr', 'hrs', 'h']:
        date = today_date - relativedelta(hours=int(x[:len(x)-1]))
        return str(date.date().isoformat())
    elif x[len(x)-1]in ['day', 'days', 'd']:
        date = today_date - relativedelta(days=int(x[:len(x)-1]))
        return str(date.date().isoformat())
    else:
        return 'unknown'
```

```python
jobs['post_date'] = jobs[4].apply(get_postdate)
# jobs[4][2]
```
